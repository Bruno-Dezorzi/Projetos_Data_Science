# California_Housing with Decision Tree

O objetivo deste script é realizar a predição dos dados do dataset california_housing usando o algortimo de árvore de decisão

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
df = fetch_california_housing(as_frame= True)
df = df.frame
display(df.head())
print(df.info())
```

```python
X_df = df.iloc[:, 1:8].values 
y_df = df.iloc[:, 8].values   
```

```python
from sklearn.model_selection import train_test_split
X_treino, X_teste,y_treino, y_teste = train_test_split(X_df, y_df, test_size=0.2, random_state= 42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
arvore = DecisionTreeRegressor()
arvore.fit(X_treino,y_treino)
```

```python
print(arvore.get_params())
```

```python
arvore.feature_importances_
```
