# Lista de 20 Exercícios em Python

## Nível Básico

### 1. Variáveis e Objetos
Crie duas variáveis `a` e `b` e atribua a elas os valores 10 e 20, respectivamente. Em seguida, troque os valores de `a` e `b` sem usar uma terceira variável.

### 2. Estrutura de Decisão
Escreva um programa que peça ao usuário para digitar um número e, em seguida, exiba se ele é positivo, negativo ou zero.

### 3. Estrutura de Repetição
Escreva um programa que imprima os números de 1 a 10 usando um laço `for`.

### 4. Listas
Crie uma lista com cinco números inteiros. Imprima o maior e o menor número da lista.

### 5. Dicionários
Crie um dicionário com os seguintes pares de chave-valor: `{'nome': 'Maria', 'idade': 25, 'cidade': 'São Paulo'}`. Exiba o valor associado à chave `idade`.

### 6. Sets
Crie dois conjuntos com os números ímpares e pares entre 1 e 10, e depois imprima a união e a interseção dos dois conjuntos.

### 7. Tuplas
Crie uma tupla com 5 elementos e tente alterar o valor do segundo elemento. O que acontece?

### 8. Estrutura de Repetição (While)
Escreva um programa que solicite um número ao usuário e continue solicitando até que ele digite um número maior que 10.

### 9. Variáveis e Objetos
Crie uma variável chamada `preco` e atribua a ela o valor 100. Em seguida, aplique um desconto de 10% e exiba o valor final.

### 10. Estrutura de Decisão (If/Else)
Escreva um programa que peça a idade do usuário e informe se ele é maior de idade ou não.

## Nível Intermediário

### 11. Estrutura de Repetição (For com Lista)
Dada uma lista de números inteiros `[2, 4, 6, 8, 10]`, escreva um programa que calcule e imprima a soma de todos os elementos da lista.

### 12. Dicionários
Crie um dicionário para armazenar o nome, a idade e o telefone de 3 pessoas. Em seguida, exiba o telefone da segunda pessoa.

### 13. Listas
Dada uma lista de números inteiros, ordene-a de forma crescente e depois exiba a lista ordenada.

### 14. Sets
Crie dois sets: um com números entre 1 e 5 e outro com números entre 3 e 7. Encontre os números que estão presentes em ambos os sets.

### 15. Tuplas
Dada uma tupla com elementos numéricos `(5, 10, 15, 20, 25)`, escreva um programa que crie uma nova tupla com o dobro de cada valor.

### 16. Estrutura de Decisão (If/Else)
Escreva um programa que solicite três números ao usuário e exiba o maior deles.

### 17. Estrutura de Repetição (For)
Crie um programa que exiba a tabuada de um número fornecido pelo usuário.

## Nível Avançado

### 18. Dicionários e Estrutura de Repetição
Crie um dicionário para armazenar as notas de 5 alunos. Em seguida, calcule e exiba a média das notas.

### 19. Estrutura de Repetição e Listas
Escreva um programa que leia 10 números inteiros e os armazene em uma lista. Depois, crie uma nova lista contendo apenas os números pares.

### 20. Variáveis, Objetos e Listas
Crie uma classe `Pessoa` com atributos nome, idade e cidade. Depois, crie uma lista com 3 objetos dessa classe e exiba os dados de cada pessoa.
